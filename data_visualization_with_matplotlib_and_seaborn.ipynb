```python
import numpy as np
x = np.linspace(0,20,100)
y = np.sin(x)

plt.figure(figsize=(6,5))
plt.plot (x,y,label= 'sin(x)',color = 'blue' )
#mandatory element
plt.title('sine wave')
plt.xlabel('x')
plt.ylabel('y')

#optional but awesome for visualization
plt.legend()
plt.tight_layout()
plt.grid(True)
plt.show()
```

```python
import numpy as np
x = np.linspace(0,20,100)
y = np.sin(x)

plt.figure(figsize=(6,5))
plt.scatter (x,y,color = 'blue' )
#mandatory element
plt.title('scatter data')
plt.xlabel('x')
plt.ylabel('y')

#optional but awesome for visualization
plt.legend()
plt.tight_layout()
plt.grid(True)
plt.show()
```

```python
x = ['cats','dogs','horse']
y = [10,3,15]

plt.figure(figsize=(6,5))
plt.bar (x,y,color = 'blue' )
#mandatory element
plt.title('bar plot')
plt.xlabel('x')
plt.ylabel('y')

#optional but awesome for visualization
plt.legend()
plt.tight_layout()
plt.grid(False)
plt.show()
```

```python
x = np.random.randn(1000)
bins = 10

plt.figure(figsize=(8,4))
plt.hist(x,bins,color = 'red' )
#mandatory element
plt.title('hist plot')
plt.xlabel('x')
plt.ylabel('y')

#optional but awesome for visualization
plt.legend()
plt.tight_layout()
plt.grid(True)
plt.show()
```

```python
x = ['cats','dogs','horse']
y = [5,10,15]

plt.figure(figsize=(3,4))
plt.pie(y,labels=x,autopct='%1.1f%%',startangle=180, colors=['red','blue','orange'])

plt.title('pie plot')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
x = np.linspace(0,20,100)
y = np.sin(x)
new = np.cos(x)
update = np.tan(x)

plt.figure(figsize=(6,5))
plt.plot(x,y,label='sin(x)',color="blue")
plt.plot(x,new,label="cos(x)",color = "red")
plt.plot(x,update,label="tan(x)",color = "orange")

plt.title("wave")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
import pandas as pd
import seaborn as sms

datasets = pd.DataFrame({
"x": ["cats","dogs","horse","car","bus","biman","rocket","space","cycle"],
"y": [1,3,2,1,3,2,1,3,2]})

plt.figure(figsize=(8,9))
sms.barplot(x="x",y="y",data = datasets,palette="viridis")

plt.title("bar plot")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```
