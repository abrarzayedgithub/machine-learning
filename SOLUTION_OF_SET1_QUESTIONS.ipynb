Write a Python script that prompts the user to enter a list of integers separated by spaces. The script
should then:

Print all the even numbers in the list.

Calculate and print the sum of all odd numbers in the list.

```python
input_list = list(map(int,input("enter a list of integers separated by spaces: ").split()))
even_numbers = [num for num in input_list if num%2 == 0]
odd_sum = sum(num for num in input_list if num % 2 !=0)

print(f"even numbers: {even_numbers}")
print(f"sum of odd numbers:{odd_sum}")
```

(b) Functions and Scope (7 Marks)
Define a function calculate_grade(scores) that takes a list of student scores (integers from 0 to 100)
and returns a dictionary with the count of grades according to the following criteria:

A: 90 and above

B: 80-89

C: 70-79

D: 60-69

F: Below 60


Demonstrate the use of local and global variables within your function.



```python
def calculate_grade(scores):
  grades = {"A":0,"B":0,"C":0,"D":0,"F":0}
  for score in scores:
    if score >= 90:
      grades["A"] += 1
    elif score >= 80:
      grades["B"] += 1
    elif score >= 70:
      grades["C"] += 1
    elif score >= 60:
      grades["D"] += 1
    else:
      grades["F"] += 1
  return grades
#example usage
student_scores = [95,88,72,67,29]
result = calculate_grade(student_scores)
print(result)
```

Given the following list of tuples representing products and their prices:


products = [("Laptop", 1200), ("Smartphone", 800), ("Tablet", 400), ("Monitor", 300)]


Perform the following tasks:

Sort the list based on price in descending order without using the built-in sort() method. (5
Marks)

Convert the list of tuples into a dictionary where the product names are keys and prices are
values. (5 Marks)

```python
products = [("Laptop", 1200), ("Smartphone", 800), ("Tablet", 400), ("Monitor", 300)]
#sort without using sort()
sorted_products = sorted(products,key=lambda x: x[1],reverse=True)
print("sorted products:", sorted_products)
 #convert to dict()
products_dict = dict(products)
print("products dictionary:",products_dict)
```

You have two dictionaries representing students enrolled in different courses:


course_python = {"Alice": 85, "Bob": 78, "Charlie": 92, "Diana": 88}

course_data_science = {"Eve": 91, "Frank": 76, "Alice": 89, "Charlie": 95}

Perform the following tasks:

Find the students enrolled in both courses using set operations. (5 Marks)

Create a new dictionary that contains the average scores of students enrolled in both courses. (5
Marks)


```python
course_python = {"Alice": 85, "Bob": 78, "Charlie": 92, "Diana": 88}

course_data_science = {"Eve": 91, "Frank": 76, "Alice": 89, "Charlie": 95}

#student enrolled in both courses
common_students = set(course_python.keys()).intersection(set(course_data_science.keys()))
print("student in both courses:",common_students)

#average scores
average_scores = {student:(course_python[student] + course_data_science[student])/2 for student in common_students}
print("average scores:",average_scores)
```

Array Creation and Manipulation (10 Marks)

Create a NumPy array A of shape (3, 5) containing random integers between 10 and 50. (2
Marks)

Compute the mean, median, and standard deviation of the array. (3 Marks)

Normalize the array so that all values are between 0 and 1.
 (5 Marks)


```python
import numpy as np
#create a random array
A = np.random.randint(10,50,(3,5))
print("Array A:",A)

#statistics
mean_val = np.mean(A)
median_val = np.median(A)
std_dev = np.std(A)
print(f"mean:{mean_val},median:{median_val},std_dev:{std_dev}")

#normalzed array
normalized_Array = (A - np.min(A))/ (np.max(A) - np.min(A))
print("normalized_Array:",normalized_Array)
```

Given a CSV file employees.csv with the following columns: 'EmployeeID' , 'Name' , 'Department' ,
'Salary' , 'JoiningDate' .

Load the dataset into a Pandas DataFrame. (2 Marks)

Display the first five rows and get basic statistics of numerical columns. (3 Marks)

Filter the DataFrame to show employees who joined after January 1, 2020, and belong to the
'Data Science' department. (5 Marks)

```python
import pandas as pd
#load dataset

df = pd.read_csv("/content/drive/MyDrive/employees.csv")

#display the five rows and statisticts
print("first five rows:",df.head())
print("basic statistics:",df.describe())

#filter data
filtered_df = df[(pd.to_datetime(df["joiningDate"])>"2020-01-02") & (df["Department"]=="Data science")]
print("filtered DataFrame:",filtered_df)
```

Using the DataFrame from part (a):

Calculate the average salary for each department. (5 Marks)

Identify the department with the highest number of employees. (5 Marks)

```python
#Calculate the average salary for each department.
avg_salary = df.groupby('Department')['Salary'].mean()
print("average salary by department:", avg_salary)

#Identify the department with the highest number of employees
dept_counts = df["Department"].value_counts()
most_employees = dept_counts.idxmax()
print(f"Department with the highest number of employees: {most_employees}")

```

Assume the 'Salary' column has missing values.

Write code to identify and count the missing values in the 'Salary' column.

Impute the missing salaries with the median salary of the respective department.

```python
#Write code to identify and count the missing values in the 'Salary' column.
missing_Salaries = df['Salary'].isnull().sum()
print(f"number of missing salaries:{missing_Salaries}")

#Impute the missing salaries with the median salary of the respective department.
df["Salary"] = df.groupby('Department')["Salary"].apply(
    lambda x: x.fillna(x.median())
 ).reset_index(level=0,drop=True)
print("missing values filled with median salary.")
```

Using the employees.csv DataFrame:

Create a bar chart showing the number of employees in each department. Include appropriate labels, title, and color customization.

```python
import matplotlib.pyplot as plt

dept_counts = df["Department"].value_counts()
dept_counts.plot(kind='bar',color='pink',title='number of employees in each department')
plt.xlabel('Department')
plt.ylabel('number of employees')
plt.show()
```

You are given a dataset sales_data.csv containing daily sales records with the following columns:

'Date' , 'ProductID' , 'QuantitySold' , 'Revenue' .

Tasks:
Load the dataset into a Pandas DataFrame. (2 Marks)

Convert the 'Date' column to datetime objects and set it as the index. (3 Marks)

Resample the data to get monthly total revenue and plot it using Matplotlib. (5 Marks)

Identify the top 3 products with the highest total sales revenue and visualize the results using a
Seaborn bar plot. (5 Marks)

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
#Load the dataset into a Pandas DataFrame.
sales_data = pd.read_csv('/content/drive/MyDrive/sales_data.csv')

#Convert the 'Date' column to datetime objects and set it as the index
sales_data["Date"] = pd.to_datetime(sales_data["Date"])
sales_data.set_index("Date",inplace=True)

#Resample the data to get monthly total revenue and plot it using Matplotlib
monthly_revenue = sales_data.resample('M')['Revenue'].sum()
monthly_revenue.plot(title='Monthly Total Revenue',ylabel='Revenue',xlabel="Month")
plt.show()
#Identify the top 3 products with the highest total sales revenue and visualize the results using a Seaborn bar plo
top_products = sales_data.groupby('ProductID')["Revenue"].sum().nlargest(3)
sns.barplot( x=top_products.index.astype(str) , y=top_products.values )
plt.title('Top 3 products by revenue')
plt.xlabel('product ID')
plt.ylabel('total revenue')
plt.show()
```

Write a Python script that performs the following:

Reads a CSV file 'weather_data.csv' containing 'Date' , 'Temperature' , 'Humidity' ,
'WindSpeed' columns. (2 Marks)

Calculates the moving average of the temperature over a 7-day window. (3 Marks)

Plots both the original temperature data and the moving average on the same graph. Include
legends and labels. (5 Marks)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

#solution for question 7
weather_data = pd.read_csv('/content/drive/MyDrive/weather_data.csv')

#moving average for temperature
weather_data['Date'] = pd.to_datetime(weather_data['Date'])
weather_data['7-Day'] = weather_data['Temperature'].rolling(window=7).mean()

#plot original and moving average
plt.figure(figsize=(11,7))
plt.plot(weather_data['Date'],weather_data['Temperature'],label='original Temperature',color='blue')
plt.plot(weather_data["Date"],weather_data['7-Day'],label='7-Day Moving Average', color='red')
plt.title('temp and 7-day moving avg')
plt.xlabel('Date')
plt.ylabel('temp')
plt.legend()
plt.show()
```
